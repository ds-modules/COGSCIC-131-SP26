```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)
```

# Create an interesting function 

Let's create a 1-D function that we'll want to find the minimum of. Let's visualize it.

```python
def target(x):
    return ((.2*x)**2)*((np.sin(x))**2 +.1)+(np.exp(.05*(x-3)))
```

```python
x=np.arange(-4,4,.01)

figure, axis = plt.subplots()
axis.plot(x,target(x))
axis.set_xlabel("x")
axis.set_ylabel("target(x)")
axis.set_title('target function to minimize');
axis.set_ylim([0.7,1.5])
axis.set_xlim([-4.2,4.2])
```

# Core idea of gradient descent

- start randomly
- follow the slope: look to the left, to the right, and move in the direction that's lower
- repeat

```python
# pick a single random starting point to try out
x0=x_start[4]

figure, axis = plt.subplots()

axis.plot(x0,target(x0),'+r')
axis.set_xlabel("x")
axis.set_ylabel("target(x)")
axis.set_title('target function to minimize');
axis.set_ylim([0.7,1.5])
axis.set_xlim([-4.2,4.2])
```

```python
# check out two points close on each side of this starting point
epsilon = .05
slope=np.empty(2)
slope[0] = x0-epsilon
slope[1] = x0+epsilon

figure, axis = plt.subplots()

axis.plot(x0,target(x0),'+r')
axis.plot(slope,target(slope),'k')

axis.set_xlabel("x")
axis.set_ylabel("target(x)")
axis.set_title('target function to minimize');
axis.set_ylim([0.7,1.5])
axis.set_xlim([-4.2,4.2])
```

```python
# Move in the direction of the gradient, unless target under then stop
if (target(slope[0])<target(slope[1])) & (target(slope[0])<target(x0)):
    x0 -=epsilon
elif (target(slope[1])<target(slope[0])) & (target(slope[1])<target(x0)):
    x0 +=epsilon
else:
    print(target(x0))
    print(target(slope))
    print('found a minimum')

slope=np.empty(2)
slope[0] = x0-epsilon
slope[1] = x0+epsilon

figure, axis = plt.subplots()
axis.plot(x0,target(x0),'+r')
axis.plot(slope,target(slope),'k')
axis.set_xlabel("x")
axis.set_ylabel("target(x)")
axis.set_title('target function to minimize');
axis.set_ylim([0.7,1.5])
axis.set_xlim([-4.2,4.2])
```

```python
x=np.arange(-4,4,.01)

# plot the true function and the final point.
figure, axis = plt.subplots()
axis.plot(x,target(x))
axis.plot(x0,target(x0),'+r')
axis.plot(slope,target(slope),'k')
axis.set_xlabel("x")
axis.set_ylabel("target(x)")
axis.set_title('target function to minimize');
axis.set_ylim([0.7,1.5])
axis.set_xlim([-4.2,4.2])
```

# What's missing?
- I was doing it by hand! We need a loop to automate this.
- It's slow! I'm moving left by a fixed $\epsilon$, no matter the slope. I should move faster when the slope is steep, slower when it's not.

Let's fix these issues below.

# adding a while loop

```python
# step size
epsilon = 0.05

# initialization
np.random.seed(42)  # for reproducibility
x_start = np.sort(np.random.uniform(-4, 4, 5))
x0 = x_start[4]

# Track the trajectory
trajectory = [x0]

# start the loop
local_min = False
iteration = 0
max_iterations = 1000  # safety limit

while not local_min and iteration < max_iterations: 
    iteration += 1
    
    # Check neighbors
    left = x0 - epsilon
    right = x0 + epsilon
    
    # Move in the direction that's lower, unless we're at a minimum
    if target(left) < target(x0) and target(left) < target(right):
        x0 = left
        trajectory.append(x0)
    elif target(right) < target(x0) and target(right) < target(left):
        x0 = right
        trajectory.append(x0)
    else:
        print(f'Found a local minimum after {iteration} iterations')
        print(f'x = {x0:.4f}, target(x) = {target(x0):.4f}')
        local_min = True

# Convert trajectory for plotting
trajectory = np.array(trajectory)

# Plot the result
figure, axis = plt.subplots()
axis.plot(x, target(x), label='Target function')
axis.plot(trajectory, target(trajectory), 'o-', color='red', markersize=4, alpha=0.5, label='Optimization path')
axis.plot(x0, target(x0), '*', color='green', markersize=15, label=f'Minimum found')
axis.set_xlabel("x")
axis.set_ylabel("target(x)")
axis.set_title(f'Gradient descent: {len(trajectory)} steps')
axis.set_ylim([0.7, 1.5])
axis.set_xlim([-4.2, 4.2])
axis.legend()
```

# Adding the actual gradient

The problem with the simple approach above is that we always move by the same step size $\epsilon$, regardless of how steep the slope is. This is inefficient: we should take bigger steps when the slope is steep, and smaller steps when we're close to the minimum.

The **gradient** tells us both the direction and the steepness of the slope. We compute it as:
$$\text{gradient} \approx \frac{f(x+\epsilon) - f(x-\epsilon)}{2\epsilon}$$

Then we update: $x \leftarrow x - \alpha \cdot \text{gradient}$

where $\alpha$ is the **learning rate** that controls how big our steps are.

```python
# learning rate (controls step size)
alpha = 0.5

# small value for computing gradient numerically
epsilon = 0.01

# convergence threshold (stop when gradient is small)
tolerance = 1e-6

# initialization - same starting point as before
np.random.seed(123)
x_start = np.sort(np.random.uniform(-4, 4, 5))
x0 = x_start[3]

# Track the trajectory
trajectory = [x0]

# start the loop
converged = False
iteration = 0
max_iterations = 1000

while not converged and iteration < max_iterations: 
    iteration += 1
    
    # Compute the gradient numerically: (f(x+eps) - f(x-eps)) / (2*eps)
    gradient = (target(x0 + epsilon) - target(x0 - epsilon)) / (2 * epsilon)
    
    # Check for convergence (gradient is close to zero)
    if abs(gradient) < tolerance:
        converged = True
        print(f'Converged after {iteration} iterations')
        print(f'x = {x0:.4f}, target(x) = {target(x0):.4f}')
    else:
        # Update x in the direction opposite to the gradient
        x0 = x0 - alpha * gradient
        trajectory.append(x0)

# Convert trajectory for plotting
trajectory = np.array(trajectory)

# Plot the result
figure, axis = plt.subplots()
axis.plot(x, target(x), label='Target function')
axis.plot(trajectory, target(trajectory), 'o-', color='red', markersize=4, alpha=0.5, label='Optimization path')
axis.plot(x0, target(x0), '*', color='green', markersize=15, label='Minimum found')
axis.set_xlabel("x")
axis.set_ylabel("target(x)")
axis.set_title(f'Gradient descent with true gradient: {len(trajectory)} steps')
axis.set_ylim([0.7, 1.5])
axis.set_xlim([-4.2, 4.2])
axis.legend()

print(f'\nCompare: fixed step needed {len(trajectory)} iterations vs gradient-based needed {iteration} iterations')
```

# Key Takeaways

1. **Gradient descent** finds a minimum by repeatedly moving in the direction opposite to the gradient (downhill).

2. **The gradient** tells us both the direction and steepness of the slope at any point.

3. **Learning rate** ($\alpha$) controls how big our steps are. Too small = slow convergence. Too large = might overshoot.

4. **Local minima**: Gradient descent finds a *local* minimum, not necessarily the *global* minimum. The result depends on where you start!

5. **Convergence**: We stop when the gradient is close to zero (flat region = minimum).
